We will create a line chart, a bar chart, and some box plots to create visualisations that will help us understand the patterns in visits to the forum and identify growth over various time ranges (yearly, monthly, etc).

The Page Views dataset can be read from [here:](https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/page_views.csv)

```python
#Read in the data into a pandas dataframe
import pandas as pd
import matplotlib.pyplot as plt

def get_data(url):
  df = pd.read_csv(url)
  print(df.head(5))
  return df


url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/page_views.csv"
df = get_data(url)
df.info()
```

The dataframe contains only 2 columns, the date (in year-month-day format) and the number of views on that date.
The date is in a string format and the view values are int64 format.


## Visualise the data in a lineplot to show the trend of page views over time.

To prepare the data for the line plot we needed to:  

*  convert the `date` column to **datetime** type 
*  create 3 new columns for `year`, `month` and `day`
*  set the index to the **date** column.
*  clean the data by filtering out days when the page views were in the top 2.5% of the dataset or bottom 2.5% of the dataset.


```python
#This code will reformat the date column and split the year, month and day out into their own columns.
from datetime import datetime, timezone
df["date"] = pd.to_datetime(df.date)
df["Year"] = df.date.dt.year
df["month_num"] = df.date.dt.month
df["Month"] = df.date.dt.month_name()
df["Day"] = df.date.dt.day

# set the index to the date column
df.set_index("date", inplace=True)
#sort the data by the value column
df_sorted = df.sort_values("value")
df_sorted.info()
```

```python
import seaborn as sns
from scipy.stats import linregress

#Removing outliers from df
def remove_outliers(df, column):
  upper_limit = df[column].quantile(0.975)
  lower_limit = df[column].quantile(0.025)
  print("Upper limit =", upper_limit,"   ", "Lower limit =", lower_limit)
  #filter all the non outlier values to give a normalised df.
  df_normal = df[(df[column]>=lower_limit)]
  df_normal = df_normal[(df_normal[column]<=upper_limit)]

  return df_normal

df_normal = remove_outliers(df_sorted, "value")
df_normal
```

```python
def display_lineplot(df, column):
  #create a subset of the df using only the net_per_10K columns and find their means.
  data = df[column]
  
  #plot the line chart of the data
  plt.figure(figsize=(18, 8))
  chart = sns.lineplot(data=data, color='lightblue')
  plt.xlabel("Date")
  plt.ylabel("Page Views")
  plt.title("Daily Free CodeCamp Forum Page Views 07/2016 - 01/2020")
  sns.set_style("whitegrid")

display_lineplot(df_normal, "value")
```

### We used seaborn to plot the line chart but used matplotlib to give us titles and figure size.

###2. Plot a bar chart 

To prepare the data for the bar chart, we created a pivot table first.


```python
import numpy as np
#To force the chart to order the month chronologically instead of alphabeitcally we use this Categroical function.
months_categories = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
df_normal["Month"] = pd.Categorical(df_normal["Month"], categories = months_categories)
df_normal.sort_values("Month")

#Create a pivot table of the page views grouped by year and plotting the months per year.
def create_pivot(df):
  pageviews_pivot = pd.pivot_table(df, values = "value", index = "Year", columns = "Month", aggfunc = np.mean )
  sns.set(rc = {'figure.figsize':(15,8)})
  pageviews_pivot.plot(kind='bar', xlabel = 'Years', ylabel = 'Average Views', title = 'Average page views by Year').legend(bbox_to_anchor=(1.2, 1))
  print(pageviews_pivot)
  return pageviews_pivot

#save series in a new variable
pageviews_pivot = create_pivot(df_normal)
```

###3.  Plot a set of box plots 

These boxplots show how the values are distributed within the years and months. Shows many outlier figures but these could be legitimate spikes in traffic to the site. 


```python
#set figure size and subplots for side by side charts
fig, axes = plt.subplots(1,2)
sns.set(rc = {'figure.figsize':(20,10)})

#set the labels of the x axis to be rotated by 45 degress for readability
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light',fontsize='large')
sns.boxplot(x= "Year", y="value", data = df_normal, width = 0.4, palette ="bright", ax=axes[0]).set(title = "Year-wise Box Plot (Trend)")
sns.boxplot(x= "Month", y="value", data = df_normal, width = 0.8, palette ="husl", ax=axes[1]).set(title ="Month-wise Box Plot (Seasonality)");
```

###4.  Now predict the number of page views in the year 2030 based on the data up to 2019
---

*  perform a linear regression and create a line of best fit 
*  predict the page views in the year 2020, 2025 and 2030 if identified trends persist.

```python
# import pandas as pd
# from scipy.stats import linregress

#This code will perform a linear regression function which the slope and intercept are 
#needed to plot line of best fit to predict future page views
def get_lineregress(df):
  x = df["Year"]
  y = df["value"]
  regression = linregress(x,y)
  
  print("Regression slope = ", regression.slope)
  print("Regression intercept =", regression.intercept)
  print("R value = ", regression[2])

  #set figure size before plotting
  plt.figure(figsize=(18, 8))
  
  #setting the years range from 2015 to a future year of 2035 and plot a prediction line using regression output
  guesstimate = np.arange(2015,2036)
  plt.plot( guesstimate, regression.intercept + regression.slope*guesstimate, 'g', label='fitted line', dashes=[20, 5, 10, 5])
  plt.title("Views per year : Prediction using line of best fit")
  plt.xlabel("Year")
  plt.ylabel("Page Views")
  return regression

regression = get_lineregress(df_normal)
```

## The line plotted above shows the predicted views that the site could have if the trend (linear regression) continued into the future. From this we can make some visual predictions or, to be more mathematical, we can use the equation y = mx + c and get estimated figures.

```python
#So using the equation for a straight line y = mx + c to predict future page views.
def page_views(slope, intercept):
  year = int(input("Please input a year (YYYY) to gain prediction: "))
  page_views = slope*year + intercept
  print("Predicted views in", year, "=", int(page_views))
  return page_views

prediction = page_views(regression.slope, regression.intercept)
```
