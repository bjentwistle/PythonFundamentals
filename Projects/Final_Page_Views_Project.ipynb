```python
#Read in the data into a pandas dataframe
import pandas as pd
import matplotlib.pyplot as plt

def get_data(url):
  df = pd.read_csv(url)
  print(df.head(5))
  return df


url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/page_views.csv"
df = get_data(url)
df.info()
```

The dataframe contains only 2 columns, the date and teh number of views on that date.
The date is in a string format and the view values are int64 format.


## Visualise the data in a lineplot to show the trend of page views over time.

To prepare the data for the line plot I need to:  

*  convert the `date` column to **datetime** type 
*  create 3 new columns for `year`, `month` and `day` (use `dt.year, dt.strftime to get the month name, dt.day`)
*  set the index to the **date** column.
*  clean the data by filtering out days when the page views were in the top 2.5% of the dataset or bottom 2.5% of the dataset.


```python
from datetime import datetime, timezone
df["date"] = pd.to_datetime(df.date)
df.dtypes

df["Year"] = df.date.dt.year
df["month_num"] = df.date.dt.month
#df["Month"] = df["date"].dt.strftime("%B")
df["Month"] = df.date.dt.month_name()
df["Day"] = df.date.dt.day

# set the index to the date column
df.set_index("date", inplace=True)

df_sorted = df.sort_values("value")
df_sorted.info()
```

```python
import seaborn as sns
from scipy.stats import linregress

#Removing outliers from df

def remove_outliers(df, column):
  upper_limit = df[column].quantile(0.975)
  lower_limit = df[column].quantile(0.025)
  print(upper_limit, lower_limit)
  #filter all the non outlier values to give a normalised df.
  df_normal = df[(df[column]>=lower_limit)]
  df_normal = df_normal[(df_normal[column]<=upper_limit)]

  return df_normal

df_normal = remove_outliers(df_sorted, "value")
df_normal
```

```python
def display_lineplot(df, column):
  #create a subset of the df using only the net_per_10K columns and find their means.
  data = df[column]
  
  #plot the chart of the data means and get labels from the years column
  plt.figure(figsize=(18, 8))
  chart = sns.lineplot(data=data, color='lightblue')
  #make a lables list from the keys of your subset df.
  # labels = df.index
  #chart.set_xticklabels( rotation=45)
  plt.title("Daily Free CodeCamp Forum Page Views 07/2016 - 01/2020")
  sns.set_style("whitegrid")

display_lineplot(df_normal, "value")
```

### We used seaborn to plot the line chart but used matplotlib to give us titles and figure size.

###2. a bar chart 

To prepare the data for the bar chart you should create a pivot table and can use pivot_table_name.plot(kind='bar')


```python
import numpy as np
#To force the chart to order the month chronologically instead of alphabeitcally we use this Categroical function.
months_categories = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
df_normal["Month"] = pd.Categorical(df_normal["Month"], categories = months_categories)
df_normal.sort_values("Month")

#Create a pivot table of the page views grouped by year and plotting the months per year.
def create_pivot(df):
  #add code below which creates a pivot table of Pop and State 
  pageviews_pivot = pd.pivot_table(df, values = "value", index = "Year", columns = "Month", aggfunc = np.mean )
  return pageviews_pivot

#save series in a new variable
sns.set(rc = {'figure.figsize':(15,8)})
pageviews_pivot = create_pivot(df_normal)
pageviews_pivot.plot(kind='bar', xlabel = 'Years', ylabel = 'Average Views', title = 'Average page views by Year').legend(bbox_to_anchor=(1.2, 1))

pageviews_pivot.head(10)
```

###3.  a set of box plots 

These box plots should show how the values are distributed within a given year or month and how it compares over time. 


```python
#To force the chart to order the month chronologically instead of alphabeitcally we use this Categroical function.
months_categories = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
df_normal["Month"] = pd.Categorical(df_normal["Month"], categories = months_categories)
df_normal.sort_values("Month")

#set figure size and subplots for side by side charts
fig, axes = plt.subplots(1,2)
sns.set(rc = {'figure.figsize':(20,10)})

#set the labels of the x axis to be rotated by 45 degress for readability
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light',fontsize='large')
sns.boxplot(x= "Year", y="value", data = df_normal, width = 0.4, palette ="bright", ax=axes[0]).set(title = "Year-wise Box Plot(Trend)")
sns.boxplot(x= "Month", y="value", data = df_normal, width = 0.8, palette ="husl", ax=axes[1]).set(title ="Month-wise Box Plot(Seasonality)");
```

###4.  predict the number of page views in the year 2020 based on the data up to 2019
---

*  perform a linear regression and create a line of best fit 
*  predict the page views in the year 2020, 2025 and 2030 if identified trends persist.

```python
# import pandas as pd
# from scipy.stats import linregress

#linear regression function needed to plot line of best fit to predict future page views
def get_lineregress(df):
  x = df["Year"]
  y = df["value"]
  regression = linregress(x,y)
  
  print("Regression slope = ", regression.slope)
  print("Regression intercept =", regression.intercept)
  print("R value = ", regression[2])

  #set figure size before plotting
  plt.figure(figsize=(18, 8))
  
  #setting the years range from 2015 to a future year of 2035 and plot using regression output
  guesstimate = np.arange(2015,2036)
  plt.plot(guesstimate, regression.intercept + regression.slope*guesstimate, 'g', label='fitted line')
  plt.title("Views per year prediction using line of best fit")
  plt.xlabel("Year")
  plt.ylabel("Page Views")
  return regression

regression = get_lineregress(df_normal)

```

## The line plotted above shows the predicted views that the site could have if the trend (linear regression) continued into the future. From this we can make some visual predictions. To be more mathematical we can use the equation y = mx + c and get estimated figures.

```python
#So using the equation for a straight line y = mx + c.
def page_views(year, slope, intercept):

  page_views = slope*year + intercept
  return page_views

prediction = page_views(2030, regression.slope, regression.intercept)
print(prediction)
```
