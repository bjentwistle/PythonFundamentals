## 1.  To get started

The function get_data() will open the data sheet, will read the rows of data and will return a list of records (dictionaries), each record containing a row of data paired with data labels.

Run this code to get the data in a form that you can work with in Python, talk about what the data looks like

```python
import csv
import urllib.request

def get_data():
    url = 'https://raw.githubusercontent.com/futureCodersSE/python-fundamentals/main/air-pollutants-daily-mean.csv'
    response = urllib.request.urlopen(url)
    lines = [l.decode('utf-8') for l in response.readlines()]
    data = csv.reader(lines)
    line_count = 0
    records = []
    for row in data:
        if line_count == 0:
            pass
        elif line_count == 1:
            keys = row
        else:
            values = row
            new_record = {k:v for k,v in zip(keys,values)}
            records.append(new_record)
        line_count += 1
    return records

air_data = get_data()
```

## 2 As a team write a function that will print the data in rows, using a **for** loop.

```python
def display_data_rows(data):
  nitric_oxcide_total = 0
  nitrogen_dioxide_total = 0
  PM10_particulate_matter_total = 0

  
  print(["Nitric Oxide"],['Nitrogen dioxide'], ['PM10 particulate matter (Grav Equiv)'] )
  
  for row [['Date'], "nitric oxide"] in data:
    if row ["Nitric oxide"] != 'nodata':
      nitric_oxcide_total = nitric_oxcide_total + float(row ["Nitric oxide"])
      nitric_oxcide_average = nitric_oxcide_total/len(data)

    if row ["Nitrogen dioxide"] != 'nodata':
      nitrogen_dioxide_total = nitrogen_dioxide_total + float(row ["Nitrogen dioxide"])
      nitrogen_dioxide_average = nitrogen_dioxide_total/len(data)

    if row ["PM10 particulate matter (Grav Equiv)"] != 'nodata':
      PM10_particulate_matter_total = PM10_particulate_matter_total + float(row ["PM10 particulate matter (Grav Equiv)"])
      PM10_particulate_matter_average = PM10_particulate_matter_total/len(data)
  
  print(nitric_oxcide_total,nitrogen_dioxide_total,PM10_particulate_matter_total)
  print(nitric_oxcide_average,nitrogen_dioxide_average,PM10_particulate_matter_average)
  


display_data_rows(air_data)
```
